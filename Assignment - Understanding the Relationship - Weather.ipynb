```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import statsmodels.formula.api as smf
import pandas as pd

# Display preferences.
%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

import warnings
warnings.filterwarnings(action="ignore")
```

```python
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

```python
weather_df.info()
```

```python
import statsmodels.api as sm

Y = weather_df['Temperature (C)']
X = weather_df[['Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Pressure (millibars)']]

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

```python
df2=weather_df.copy()
```

```python
df2["target"]=df2["Apparent Temperature (C)"]-df2["Temperature (C)"]
df2.head()
```

```python
Y2 = df2['target']
X2 = df2[['Humidity', 'Wind Speed (km/h)']]

X2 = sm.add_constant(X2)

results2 = sm.OLS(Y2, X2).fit()

results2.summary()
```

---

### R-squared value is decreased from 0.421 to 0.288, which means that our second model is performing worse than the first one. This tells us that only 2 explonatory variable are not enough to create a model to predict the targeted value.

---


```python
df2["hum_windspeed"]=df2["Humidity"]*df2["Wind Speed (km/h)"]
df2.head()
```

```python
Y2 = df2['target']
X2 = df2[['Humidity', 'Wind Speed (km/h)', 'hum_windspeed']]

X2 = sm.add_constant(X2)

results2 = sm.OLS(Y2, X2).fit()

results2.summary()
```

---

### R-squared value is increased from 0.288 to 0.341, so we have a better model now. Coefficints of Humidity and Wind Speed have changed signs from negative to positive. Also the value of all coefficients, including the constant, has became smaller.
